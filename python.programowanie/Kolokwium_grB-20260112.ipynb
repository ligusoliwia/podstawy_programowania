#### Kryteria oceniania
Ocenie podlegają następujące elementy:
* poprawność proponowanego rozwiązania
* kompletność wykonania zadania
* czytelność kodu
* odpowiednie oznaczanie typów
* zabezpieczenie kodu przed niepoprawnym użyciem
* łatwość obsługi programu.

### Zadanie 9. (5 pkt.)
Napisz klasę `Temperatura`. Jako parametr przyjmuje liczbę rzeczywistą nieujemną będącą temperaturą wyrażoną w Kelvinach. Aktualizacja temperatury odbywa się poprzez użycie odpowiedniej metody jako atrybutu. Obiekt klasy przechowuje temperaturę. Klasa implementuje następujące metody:
* celsjusz - konwertuje aktualną temperaturę na stopnie Celsjusza i zwraca ją do programu, metoda obsługiwana jak atrybut. Po przypisaniu nowej wartości konwertuje ją na Kelviny i aktualizuje wartość aktualnej temperatury.

Podpowiedź: 
$$T_C = T_K - 273.15$$
gdzie:
- $T_C$ - temperatura wyrażona w stopniach Celsjusza,
- $T_K$ - temperatura wyrażona w Kelvinach.

Podpowiedź: $0$ Kelvinów oznacza zero absolutne i w rozumieniu fizyki klasycznej temperatura nie może osiągnąć wartości poniżej zera absolutnego.
