### Odczytywanie plików tekstowych
Do zestawu podstawowych operacji w Pythonie należy wczytywanie i obsługa plików tekstowych. Do otwierania plików służy funkcja `open()`. Jest to funkcja która może przyjmować wiele parametrów. Omówimy tylko kilka z nich.

Podstawowym parametrem jest ścieżka do pliku jaki funkcja ma otworzyć. Może to być ścieżka bezwzględna, np.: */home/user/podstawy_programowania/plik.txt*, ale może to być również ścieżka względna, np.: *../podstawy_programowania/plik.txt*. W systemie Windows ścieżki jako separator domyślnie wykorzystują tzw. *backslash*, czyli "\\". Niestety, w Pythonie *backslash* jest wykorzystywany do tworzenia znaków specjalnych w tekście jak znak nowej linii ("\\n") czy tabulator ("\\t"). W przypadku takich ścieżek należy więc każdy *backslash* w ścieżce zamienić na podwójny *backslash*, czyli zastąpić taki tekst *C:\\plik.txt* takim tekstem *C:\\\\plik.txt*. Można również zastąpić *backslash* zwykłym *slash*-em co jest o wiele wygodniejsze.

Umiejętność obsługi tego podstawowego parametru funkcji `open()` pozwala otworzyć plik w trybie do odczytu i pobierać jego treść. Funkcja `open()` zwraca **obiekt** pliku, czyli coś w rodzaju uchwytu do tego pliku, który pozwala na wykonywanie operacji na tym pliku. Obiekt ten posiada zestaw metod, które można wykorzystać do wykonywania operacji na pliku. Podstawowa operacja to odczytywanie całej zawartości pliku o nazwie `.read()`. Zawartość takiego pliku jest zwracana w postaci zwykłego tekstu i w ten sposób należy go następnie traktować. W przypadku dużych plików mamy do dyspozycji metodę `.readline()`. Pozwala ona wczytywać kolejne linijki tekstu z pliku. Mamy też metodę `.readlines()`, która wczytuje całą treść pliku i umieszcza ją na liście, gdzie każda linijka to oddzielny element.

Po każdym otwarciu pliku należy bezwzględnie pamiętać o tym aby został on zamknięty. W przeciwnym przypadku plik pozostanie oznaczony jako otwarty i żaden inny program nie będzie mógł go zedytować. Jedynym praktycznym sposobem na wykasowanie takiego rejestru jest restart systemu. Aby zamknąć plik należy użyć jego metody `.close()`. Powoduje ona usunięcie rejestru o otwarciu pliku, odpięcie uchwytu w Pythonie od tego pliku oraz usunięcie zmiennej tego pliku z pamięci.

```Python
plik = open('plik.txt')
tresc_pliku = plik.read()

plik.close()
```

Należy pamiętać o tym, że pojedynczy uchwyt do pliku jest traktowany jak obiekt iterowalny. Niesie to za sobą dwie konsekwencje. Pierwsza jest taka, że nie można dwa razy odczytać tego samego fragmentu pliku. Oznacza to, że poniższy kod jest niepoprawny z punktu widzenia programisty.
```Python
plik = open('plik.txt')
tresc_pliku = plik.read()
linijki = plik.readlines()
plik.close()
```

Efekt będzie taki, że w zmiennej `tresc_pliku` znajdzie się cała zawartość tego pliku, ale zmienna `linijki` pozostanie pusta, ponieważ nie zostało w pliku już nic do odczytania.

Druga konsekwencja jest taka, że można traktować plik tak samo jako listę i w przypadku odczytywania sekwencyjnego zamiast stosować taką konstrukcję pętli:
```Python
plik = open('plik.txt')
while True:
    linia = plik.readline()
    if linia == '':
        break
    # operacje na zmiennej linia
plik.close()
```
gdzie kłopotliwe jest sprawdzanie czy plik się skończył i nie zabezpiecza programu przed sytuacją kiedy plik po prostu zawiera pustą linijkę. Zamiast tego można zastosować pętlę `for`:
```Python
plik = open('plik.txt')
for linia in plik:
    #operacje na zmiennej linia
plik.close()
```
Tutaj kod jest o wiele krótszy, unikamy własnoręcznego sprawdzania czy plik się skończył i nie musimy się przejmować czy w pliku występują puste linijki.

### Zapisywanie plików tekstowych
Funkcja `open()` posiada również parametr pozwalający określić w jakim trybie plik ma zostać otwarty. Trybem domyślnym jest tryb odczytu. Dlatego nie musieliśmy dodawać tego parametru w programach, które pisaliśmy do tej pory. Wyróżniamy 3 tryby otwarcia pliku:
* tryb odczytu ("r") - pozwala na odczytywanie zawartości pliku. Jeżeli plik o podanej nazwie nie istnieje to zwraca błąd;
* tryb zapisu ("w") - pozwala na zapisywanie zawartości do pliku. Jeżeli plik o podanej nazwie nie istnieje to tworzy nowy plik. Jeżeli w pliku znajduje się jakakolwiek zawartość to **jest ona wymazywana**;
* tryb dopisywania ("a") - pozwala na zapisywanie zawartości do pliku. Jeżeli plik o podanej nazwie nie istnieje to tworzy nowy plik. Jeżeli w pliku znajduje się jakakolwiek zawartość to **nowa treść jest dopisywana na końcu**.

Dodatkowo, do każdego trybu można dodać "b" co oznacza, że plik będzie obsługiwany w trybie binarnym, jednak jest to zagadnienie do samodzielnego zgłębienia i nie będziemy go szerzej omawiać.

Kiedy zależy nam na tym aby plik jednocześnie obsługiwał odczyt i zapis to można zastosować następujące tryby:
* tryb odczyt i zapis ("r+") - pozwala na odczyt i zapis do pliku. Jeżeli plik nie istnieje to zwraca błąd;
* tryb zapis i odczyt ("w+") - pozwala zapisać treść do pliku, a następnie ją odczytać. Jeżeli plik nie istnieje to tworzony jest nowy. Jeżeli w pliku znajduje się jakakolwiek zawartość to **jest ona wymazywana**;
* tryb dopisywania i odczytu ("a+") - pozwala zapisać treść do pliku, a następnie ją odczytać. Jeżeli plik nie istnieje to tworzony jest nowy. Jeżeli w pliku znajduje się jakakolwiek zawartość to **nowa treść jest dopisywana na końcu**.

Aby otworzyć plik w ustalonym trybie należy jako drugi parametr funkcji `open()` podać nazwę trybu. Przykładowo, jeśli chcemy otworzyć plik w trybie do zapisu to podajemy "w":
```Python
plik = open('plik.txt',"w")
# zapisywanie treści pliku
plik.close()
```

Do zapisywania tekstu w pliku służy metoda `.write()`, która jako parametr przyjmuje tekst jaki chcemy zapisać do pliku. Należy jednak pamiętać, że ta metoda nie tworzy nowej linii. Ona zawsze dopisuje na końcu tekst poprzedniego, na przykład:
```Python
plik = open('plik.txt',"w")
plik.write('Ala ma kota')
plik.write('Kot ma Alę')
plik.close()

plik = open('plik.txt','r')
print(plik.read())
plik.close()
```
```Python
Out:
Ala ma kotaKot ma Alę
```
Należy zwrócić tutaj uwagę, że wypisany tekst został sklejony. Wynika to z faktu, że nigdzie nie zapisaliśmy podziału linii. Na końcu pierwszej linijki powinniśmy dodać znak nowej lini *\n*:
```Python
plik = open('plik.txt',"w")
plik.write('Ala ma kota\n')
plik.write('Kot ma Alę')
plik.close()

plik = open('plik.txt','r')
print(plik.read())
plik.close()
```
```Python
Out:
Ala ma kota
Kot ma Alę
```

Kiedy zapisujemy zawartość do pliku to w rzeczywistości Python nie zapisuje zmian w pliku natychmiast, ale zbiera je w buforze żeby zaoszczędzić liczbę operacji wejścia/wyjścia na dysku. Z tego powodu zmiany w pliku nie są widoczne natychmiast po wykonaniu metody `.write()`, a dopiero po zamknięciu pliku. Jeżeli chcemy widzieć zmiany w pliku jeszcze przed zamknięciem pliku i jednocześnie chcemy się zabezpieczyć przed utratą danych w przypadku np. błędu naszego programu to po każdym wywołaniu metody `.write()` należy wywołać metodę `.flush()`. Nie przyjmuje ona żadnych parametrów i powoduje opróżnienie bufora, przez co oczekujące zmiany są zapisywane na dysku twardym.

### Konteksty
Potrzeba zamykania pliku w określonym miejscu w kodzie naszego programu jest kłopotliwe. Wymusza na programiście konieczność pamiętania o jednej instrukcji, która z punktu widzenia programu nie robi nic przydatnego. Stąd często zapomina się o dodawaniu tej instrukcji, a to powoduje narastające problemy w systemie plików na dysku twardym. Dodatkow, nawet jeśli programista pamiętał o dodaniu tej instrukcji w programie to nadal może się ona nie wykonać ze względu na wyrzucony wyjątek co powoduje przerwanie wykonywania programu lub też ze względu na zwyczajną pomyłkę i umieszczenie instrukcji w nieodpowiednim miejscu w przypadku bardziej skomplikowanych programów.

Wygodnie byłoby gdyby coś dbało o zamknięcie otwartego pliku automatycznie. W Pythonie istnieje mechanizm tzw. **kontekstów**. Kontekst to pewien wydzielony fragment programu, który posiada odwołanie do określonego obiektu i w kontekście tego obiektu wykonuje określone instrukcje, a następnie po zakończeniu kontekstu obiekt ten znika. Mechanizm ten jest wkorzystywany w wielu przypadkach jednak podstawowym jest właśnie obsługa plików. 

Do utworzenia kontekstu służy słowo kluczowe `with`, następnie musi nastąpić inicjalizacja obiektu, na którym kontekst ma operować (np. otwarcie pliku). Następnie należy umieścić słowo kluczowe `as` i po nim nazwę nazwę zmiennej, do której chcemy przypisać obiekt. Na koniec wstawiamy dwukropek i od nowej linii z odpowiednim wcięciem piszemy instrukcje wykonywane w kontekście:
```Python
with inicjalizacja_obiektu as zmienna:
    #instrukje wykonywane na zmiennej zmienna
```

Przykładowo, jeśli chcielibyśmy otworzyć znany nam już plik *plik.txt* w kontekście to użylibyśmy następującego kodu:
```Python
with open('plik.txt') as plik:
    tekst = plik.read

print(tekst)
```
Widać tutaj, że plik został otwarty w kontekście, a następnie jego treść zostałą wczytana w całości do zmiennej tekst. Na koniec, już poza kontekstem tekst został wypisany na ekran. Zmienne utworzone w kontekście mają zasięg również poza kontekstem, ale obiekt, który jest inicjalizowany jako kontekst kończy swoje istnienie w momencie zakończenia kontekstu. W tym przypadku otwarty plik po zakończeniu kontekstu został zamknięty i nie jest już dłużej dostępny.
