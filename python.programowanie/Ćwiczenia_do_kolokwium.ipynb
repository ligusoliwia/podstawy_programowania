#### Zadanie 4.

#### Zadanie 5.

#### Zadanie 6. 
Napisz program, który oblicza $n$ liczb doskonałych i wypisuje je na ekran. 

Podpowiedź: liczba doskonała to taka, która jest równa sumie swoich dzielników właściwych:
$$n=6$$
$$\text{dzielniki liczby 6} = \{1,2,3\}$$
$$1+2+3=6$$

Podpowiedź 2: liczby doskonałe są niezwykle rzadkie. Wyznaczenie więcej niż 3 liczb doskonałych spowoduje zawieszenie programu.

#### Zadanie 10.

#### Zadanie 14.
Napisz klasę `Temperatura`. Jako parametr przyjmuje liczbę rzeczywistą nieujemną będącą temperaturą wyrażoną w Kelvinach. Aktualizacja temperatury odbywa się poprzez użycie odpowiedniej metody jako atrybutu. Obiekt klasy przechowuje temperaturę. Klasa implementuje następujące metody:
* celsjusz - konwertuje aktualną temperaturę na stopnie Celsjusza i zwraca ją do programu, metoda obsługiwana jak atrybut. Po przypisaniu nowej wartości konwertuje ją na Kelviny i aktualizuje wartość aktualnej temperatury.

Podpowiedź: 
$$T_C = T_K - 273.15$$
gdzie:
- $T_C$ - temperatura wyrażona w stopniach Celsjusza,
- $T_K$ - temperatura wyrażona w Kelvinach.
* fahrenheit - konwertuje aktualną temperaturę na stopnie Celsjusza i zwraca ją do programu, metoda obsługiwana jak atrybut. Po przypisaniu nowej wartości konwertuje ją na Kelviny i aktualizuje wartość aktualnej temperatury.

Podpowiedź:
$$T_F = T_C * \frac{9}{5} + 32$$
gdzie:
- $T_F$ - temperatura wyrażona w stopniach Fahrenheita,
- $T_C$ - temperatura wyrażona w stopniach Celsjusza.

Utwórz obiekt tej klasy i przetestuje przypisywanie i odczytywanie temperatur.

Podpowiedź: $0$ Kelvinów oznacza zero absolutne i w rozumieniu fizyki klasycznej temperatura nie może osiągnąć wartości poniżej zera absolutnego.

### Zadanie na ocenę 5

Napisz klasę `Zwierze`. Klasa przyjmuje dwa parametry:
* imie: tekst długości od 3 do 20 znaków, oznacza imię zwierzęcia
* glos: tekst długości od 3 do 10 znaków, oznacza odgłos jaki wydaje to zwierzę
Klasa wymusza implementację metody `przywolaj()` w klasach potomnych.

Napisz klasę `Pies` dziedziczącą z klasy `Zwierze`. Klasa przyjmuje parametr imie oraz glos z wartością domyślną *Hau!*. Metoda `przywolaj()` powoduje najpierw szczeknięcie psa (na ekran wypisany jest atrybut glos) a następnie pies przychodzi (na ekran wypisywany jest napis *Pies przyszedł*).

Napisz klasę `Kot` dziedziczącą z klasy `Zwierze`. Klasa przyjmuje parametr imie oraz glos z wartością domyślną *Miau!*. Kot posiada dodatkowy atrybut, który domyślnie jest równy $0$. Metoda `przywolaj()` najpierw dokonuje inkrementacji dodatkowego atrybutu, następnie sprawdza czy wartość tego atrybutu jest podzielna przez długość imienia kota. Jeżeli tak to kot miauczy (na ekran wypisany jest atrybut glos), a następnie kot przychodzi (na ekran wypisywany jest napis *Kot przyszedł*). Jeżeli wartość tego atrybutu nie jest podzielna przez długość imienia to kot nie przychodzi (nic nie jest wypisywane na ekran).

Napisz klasę `Czlowiek` dziedziczącą z klasy `Zwierze`. Klasa przyjmuje tylko parametr imie. Metoda `przywolaj()` przyjmuje dodatkowy parametr którym jest obiekt zwierzecia. Wywołanie metody powoduje, że człowiek woła zwierze (na ekran wypisywane jest imię zwierzęcia z wykrzyknikiem na końcu), następnie zwierzę jest przywoływane (na przekazanym obiekcie wywoływana jest metoda `prywolaj()` tego obiektu). 

Utwórz obiekty klas `Pies`, `Kot` i `Czlowiek`. Zaprezentuj przywoływanie zwierzęcia. Kota wołaj aż do skutku.

Podpowiedź: Jeśli cały kod zostanie zapisany w jednej komórce to po każdym uruchomieniu obiekty zostaną utworzone na nowo i wymazane zostaną ich aktualne parametry. Zamiast tego obiekty należy utworzyć w jednej komórce, a wywoływanie metod należy testować w innych komórkach.
