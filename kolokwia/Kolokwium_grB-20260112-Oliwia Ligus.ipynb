### Zadanie 6. (5/5 pkt.)
Napisz klasę `Samochod`. Klasa przyjmuje następujące parametry:
* marka - tekst długości od 3 do 20 znaków
* kolor - tekst o możliwych wartościach: *czerwony*, *czarny*, *niebieski*, *szary*.
* predkosc - liczba naturalna o maksymalnej wartości $400$

Klasa implementuje następujące metody:
* jedz() - wypisuje na ekran informację o tym, że samochód się przemieszcza
* stoj() - wypisuje na ekran informację o tym, że samochód się zatrzymał

Dodatkowo, przeciąż metodę \_\_str\_\_() aby po wypisaniu obiektu wypisany został napis: *Pojazd {marka} w kolorze {kolor} o prędkości maksymalnej {predkosc} km/h*.

````python
class SAMOCHOD:
    def __init__(self, marka: str, kolor: str, predkosc: int):
        if 3 <= len(marka) <= 20:
            self.marka = marka
        kolorki = ['czerwony', 'czarny', 'niebieski', 'szary']
        if kolor in kolorki:
            self.kolor = kolor
        if 0 <= predkosc <= 400:
            self.predkosc = predkosc
    def jedz(self):
        return 'zielone światło - auto jedzie!'
    def stoj(self):
        return 'czerwone światło - auto stoi!'
    def __str__(self):
        return f'pojazd {self.marka} w kolorze {self.kolor} o prędkości maksymalnej {self.predkosc} km/h'

auto = SAMOCHOD('porsche', 'czarny', 320)
print(auto)
print(auto.jedz) # tutaj zabrakło nawiasów - jest to metoda - ale nie jest to elementem zadania więc nie obcinam punktów
print(auto.stoj)
````

### Zadanie 7. (4/5 pkt.)
Napisz klasę `Punkt`. Jako parametry przyjmuje współrzędne punktu na płaszczyźnie. Posiada dwie metody:
* przesun(dx, dy) - przyjmuje wektor, przesuwa punkt o podany wektor,
* odleglosc(inny) - jako parametr przyjmuje obiekt innego punktu, oblicza odległość bieżącego punktu od podanego w parametrze. Wykorzystaj odległość Manhattan:
$$d_M=\sum_{i=1}^n|x_i-y_i|$$

Aby obliczyć wartość bezwzględną wykorzystaj funkcję `abs()`

Utwórz dwa obiekty tej klasy i oblicz odległość między nimi.

````python
class PUNKT:
    def __init__(self, x: int, y: int):
        self.x = x
        self.y = y
    def przesun(self, dx: int, dy: int):
        self.x += dx
        self.y += dy
        return self.x, self.y # tego nie było w zadaniu, ale zasadniczo nie przeszkadza
    def odleglosc(self, inny):
        inny = abs(self.x - self.y) # w tym momencie nadpisała sobie pani przekazany parametr
        # d = abs(self.x - inny.x) + abs(self.y - inny.y) # w taki sposób należałoby obliczyć odległość
        return inny # zwracamy odległość, nie obiekt
    
p1 = PUNKT(2, 4)
p2 = PUNKT(3, 5)
p1.przesun(2, 2)
p2.przesun(1, 1)
p1.odleglosc(p2)
````

Napisz klasę `Licznik`. Jako parametr przyjmuje element startowy będący liczbą całkowitą - domyślnie $0$ oraz skok będący liczbą naturalną. Klasa posiada następujące metody:
* inkrementuj() - zwiększa wartość licznika o skok,
* dekrementuj() - zmniejsza wartość licznika o skok,
* resetuj() - przywraca wartość licznika do wartości początkowej,
* wartosc - metoda obsługiwana jako atrybut, bez możliwości edycji, zwraca aktualną wartość licznika.

````python
class LICZNIK:
    def __init__(self, liczba = 0):
        self.liczba = liczba
    def inkrementuj(self, skok):
        self.liczba += skok
        return self.liczba
    def dekrementuj(self, skok):
        self.liczba -= skok
        return self.liczba
    def resetuj(self):
        self.liczba = 0 # a skąd pani wie, że liczba początkowa to 0?
        return self.liczba

n = LICZNIK(5)
print(n.inkrementuj(3))
print(n.dekrementuj(2))

# założenie było takie, że skok jest przekazywany do konstruktora i inkrementacja oraz dekrementacja powoduję zmianę wartości o stałą wartość
````
